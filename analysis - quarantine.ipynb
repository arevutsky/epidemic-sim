```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

names = ["q01", "q02", "q03"]
source = []
for name in names:
    source.append(pd.read_csv("data/epidemic-" + name + ".csv"))
a = np.array(list(range(0,len(source[0]["infections"]))))
x = a.astype(float)

```

```python
for s, n in zip(source, names):
    plt.plot(x, s["infections"], label="q="+n)
    
plt.legend()
plt.show()
```

```python
for s, n in zip(source, names):
    plt.plot(x, s["deaths"], label="q="+n)
    
plt.legend()
plt.show()
```

```python
for s, n in zip(source, names):
    plt.plot(x, s["infecteds"], label="q="+n)
    
plt.legend()
plt.show()
```
